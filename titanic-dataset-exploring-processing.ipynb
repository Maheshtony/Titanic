# Grouping and aggrigations


```python
# to group 2 columns we use groupby function. Below code gives you the age 

train.groupby('Sex').Age.median()
```

```python
# groupby for Passaner class and fare
train.groupby(['Pclass']).Fare.median()
```

```python
# we can groupby Pclass for 2 different variables such as Fare and age as shown below
train.groupby(['Pclass'])['Fare', 'Age'].median()
```

```python
# using aggregation function we can get more that 2 variables in the agg by creating a dictionary

train.groupby(['Pclass']).agg({'Fare' : 'mean', 'Age' : 'median'}) 
```

```python
# how to create a Complecated aggregation

aggregations = {
    'Fare' : { # Working on fare column
            'mean_Fare' : 'mean',  # mean fare
        'median_Fare' : 'median', # we will get the median
        'max_Fare' : max,
        'min_Fare' : np.min
    },
    
    'Age' : {  # working on Age column
            'meadian_Age' : 'median',
            'min_Age' : min,
            'max_Age' : max,
            'range_Age' : lambda x: max(x) - min(x)
    }
 }


```

```python
# we sent aggregation dictionary into agg function you can ignore the warning
train.groupby(['Pclass']).agg(aggregations)
```

```python
train.groupby(['Pclass', 'Embarked']).Fare.median()
```
